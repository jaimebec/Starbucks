# Description

The dataset includes nutritial information for starbucks' food and drink menu items.

Beberage_category: the category of the beberage

Beberage: The name of the drink/food

Beberage_prep: alternative ingredients in the drink/food

Trans fat (g): Amount of trans fat in grams in the drink/food

Saturated fat (g): Amount of saturated fat in grams in the drink/food

Total Carbohydrates (g): Amount of carbohydrates in the drink/food

Cholesterol (mg): amount of Cholesterol in miligrams in the drink/food

Protein: Amount of protein in grams in the drink/food

Sodium (mg): Amount of sodium in miligrams the drink/food

Sugars (g): Amount of sugar in grams in the drink/food

Caffeine (mg): Amount of caffeine in the drink/food

# 1. Libraries and Data Loading

```python
# Data Manipulation
import pandas as pd

# Statistics


# Data vizualitation
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

sns.set_palette("husl")
```

```python
# load database
menu = pd.read_csv('starbucks_drinkMenu_expanded.csv')

# dropping unwanted columms 
menu.drop([' Dietary Fibre (g)', 'Vitamin A (% DV) ', 'Vitamin C (% DV)', ' Calcium (% DV) ', 'Iron (% DV) ', ' Sodium (mg)'], axis=1, inplace=True)

menu.columns
```

## Summary of data

```python
menu[' Total Fat (g)'] = pd.to_numeric(menu[' Total Fat (g)'], errors='coerce')
menu['Caffeine (mg)'] = pd.to_numeric(menu['Caffeine (mg)'], errors='coerce')

print(menu.info())

print(menu.describe())

print(menu.shape)

menu.head(7)
```

# 2. Descriptive Analysis

## Quantitative Variables
### Continous
Calories, Total Fat, Trans Fat, Saturated Fat, Total Carbohydrates(g), Cholesterol (mg), Sugar (g), Protein (g), Caffeine (mg)

```python
def distribution(column, binN, axis):
    binned = pd.cut(menu[column], binN, right=False)
    
    Df = pd.DataFrame({'Range': binned.value_counts().index, 'Frequency':binned.value_counts().values})
    Df['Percent'] = Df['Frequency'] / Df['Frequency'].sum() * 100
    Df = Df.sort_values('Range')

    ax = menu[column].plot.hist(figsize=(14, 14), bins=binN, ax=axis, edgecolor ='black')
    ax.set_xlabel(column)
    
    return Df
    
menu.columns
```

```python
# Calories

fig = plt.figure()
ax1 = fig.add_subplot(3, 3, 1)
ax2 = fig.add_subplot(3, 3, 2)
ax3 = fig.add_subplot(3, 3, 3)
ax4 = fig.add_subplot(3, 3, 4)
ax5 = fig.add_subplot(3, 3, 5)
ax6 = fig.add_subplot(3, 3, 6)
ax7 = fig.add_subplot(3, 3, 7)
ax8 = fig.add_subplot(3, 3, 8)

distribution('Calories', 20, ax1)
distribution(' Total Fat (g)', 15, ax2)
distribution('Saturated Fat (g)', 5, ax3)
distribution(' Total Carbohydrates (g) ', 20, ax4)
distribution('Cholesterol (mg)', 15, ax5)
distribution(' Sugars (g)', 15, ax6)
distribution(' Protein (g) ', 15, ax7)
distribution('Caffeine (mg)', 15, ax8)

plt.rc('figure', figsize=(15, 15))
```

From the above distributions a couple of conclusions could be said

Calories 
- Most of the products have between 200 and 400 calories
- There are a lot of outliers to the left, which correspond to the low-calories alternatives of starbucks

Fat
- Most of the products are low-fat

Carbohydrates
- Most of the products have between 100 and 200 carbohydrates

Sugars
- Most of the products have between 20 and 40 grams of sugar

Protein
- Most of the product are low-protein


## Qualitative Variables
### Nominal
Beverage_Category, Beverage, Category

```python
# Beverage

menu['Beverage']
menu['Beverage_category']
menu['Beverage_prep']

menu.columns
```

```python

def qualitative(column):
    
    Df = pd.DataFrame({'Type': menu[column].value_counts().index, 'Frequency':menu[column].value_counts().values})
    Df['Percent'] = Df['Frequency'] / Df['Frequency'].sum() * 100
    Df = Df.sort_values('Frequency', ascending=False)
    print(Df)
    
    plt.rc('figure',figsize=(7,7))
    fig, ax = plt.subplots()
    ax.bar(Df['Type'], Df['Frequency'])
    ax.set_xticklabels(Df['Type'], rotation=90)
    plt.show()
```

```python
qualitative('Beverage')
```

Tazo Full.Leaf Red Tea Latte, Caffe Latte and white chocolate mocha are the products with most variations in the menu

```python
qualitative('Beverage_category')
```

Classic Espresso Drinks, Tazo Tead Drinks and Signature Espresso Drinks are the categories with the most products.

```python
qualitative('Beverage_prep')
```

Soymilk, 2% Milk and Grande Nonfat Milk are the most frequent preparations.

# 3. Exploratory Analysis

On qualitative Fields: 'Beverage_category', 'Beverage', 'Beverage_prep'


## 3.1 Analysis of Beverages categories with most/least calories and protein

```python
def tqualitative(qualitative, quantitative, N=10, least=True):
    BeverageGroup = menu.groupby(qualitative)[quantitative].mean().sort_values(ascending=least)[:N]

    print(BeverageGroup)

    x = sns.barplot(x=qualitative, y=quantitative, data=menu, order=BeverageGroup.index)
    plt.xticks(rotation=90) 
    #plt.show()
    return x
```

```python
# categries with the most calories on average

plt.rc('xtick', labelsize=9) 
y = tqualitative('Beverage_category', 'Calories', 10, False)
```

```python
# categories with the most protein on average

tqualitative('Beverage_category', ' Protein (g) ', 10, False)
```

## 3.2 Analysis of Beverage with most/least calories and protein

```python
# top 10 beverages with most calories

tqualitative('Beverage', 'Calories', 10, False)
```

```python
# top 10 beverages with least calories

tqualitative('Beverage', 'Calories', 10, True)
```
